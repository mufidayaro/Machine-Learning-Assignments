```python
import matplotlib.pyplot as plt
import seaborn as sns

pumpkin_data = sns.load_dataset('penguins')

plt.figure(figsize=(10, 6))
plt.scatter(pumpkin_data['bill_length_mm'], pumpkin_data['flipper_length_mm'])
plt.xlabel('Bill Length (mm)')
plt.ylabel('Flipper Length (mm)')
plt.title('Scatter Plot of Bill Length vs Flipper Length')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='body_mass_g', data=pumpkin_data)
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.title('Boxplot of Body Mass by Species')
plt.show()
```
